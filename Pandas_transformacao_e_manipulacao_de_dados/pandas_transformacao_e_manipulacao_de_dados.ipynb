Fazendo a importação dos dados e das bibliotecas que serão usadas no curso

```python
#importando bibliotecas e dados
import pandas as pd
import numpy as np


dados = pd.read_json('https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_hospedagem.json')
dados
```

---
Correção de estrutura do DataFrame
---

Após importar os dados percebe-se que os mesmos não estão normalizados pois a coluna "info_moveis" apresenta muitos dados agrupados. Para isso será utilizado o método *json_normalize*

```python
#Normalizando os dados JSON
dados = pd.json_normalize(dados['info_moveis'])
dados
```

A visualização dos dados melhorou, porém ainda há dados agrupados nas linhas só que me formato de lista. Para podermos espandir esses dados usaremos o método *explode()*.

O mesmo de acordo com a documentação necessita de da informação da coluna que será "explodida". Por isso será extraído o nome das colunas em uma lista para ser passada como parâmtro para o *explode()*

```python
#Extraindo as colunas em forma de lista
colunas = list(dados.columns)
colunas
```

```python
#"Explodindo" colunas com listas internas nas linhas
dados = dados.explode(colunas[3:])
dados
```

---
Depois será necessário resetar o index pois o numero de linhas aumentou depois que os dados foram expandidos

```python
#Resetando o index
dados.reset_index(inplace=True, drop=True)
dados.head()
```

---
após expandir os dados das colunas é necessário fazer a manuntenção das mesmas. Deve-sem obervar as colunas que são do Dtype numericas (int, float) e regularizá-las de acordo.

Para isso será verificado visualmente essa informação e logo após utilizado o método *astype()* para a conversão.

```python
#pegando informaçeos gerais das colunas
dados.info()
```

```python
#Alterando dtype de coluna para int64
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)

dados.info()
```

```python
#Fazendo alteração de colunas para int (porém várias dessa vez)
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']

dados[col_numericas] = dados[col_numericas].astype(np.int64)
dados.info()
```

```python
#alterando colunas para float
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)

dados.info()
```

A coluna "preco" é uma das que possuem caracteres especiais. Logo se for aplicada o *astype()* nela obteremos:

```python
dados['avaliacao_geral'] = dados['preco'].astype(np.float64)
```

Para isso será necessária o tratamento dos dados que estão nessa coluna em particular.

Para tratar esses dados usaremos o método *replace()* para retirar os "$" e as vírgulas para que fique em um padrão que o *astype()* possa identificar. Depois disso será feito a conversão

Como teremos que usar em cada linha da coluna, usaremos o *lambda*.

```python
#Aplicando uma limpeza de dados na coluna 'preco'
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',','').strip())
dados['preco']
```

```python
#fazendo alteração da coluna para float
dados['preco'] = dados['preco'].astype(np.float64)
dados.info()
```

A mesma limpesa será feita em outras colunas como "taxa_deposito" e "taxa_limpeza". Porém ao invés de ser aplicada no formato de series como da ultima vez, será usado o formato de dataframe.

Para isso usaremos o método *applymap()* ao invés do *apply()* comum

```python
#Fazendo a limpeza dos dados de outras colunas
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',','').strip())
dados[['taxa_deposito', 'taxa_limpeza']]
```

```python
#Transformando os dados para float64
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
dados.info()
```

---
Manipulando Texto
---

Será necessário agora tratar uma coluna específica chamada "descrica_local", a qual possui informações sobre o a descrição do imóvel. Para isso utilizaremos o REGEX e os métodos de string disponíveis. Sérá feito:

1. Retirada dos carcteres especiais;
2. Remoção de hífens fora das palavras compostas ou fora do padrão.

```python
#transformando os dados em string e passando todas as letras para minúsculo
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados['descricao_local']
```

```python
#Retirada dos caracteres especiais
dados['descricao_local'] = dados['descricao_local'].replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_local']
```

```python
#remoção de hífens fora das palavras compostas
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
dados['descricao_local']

#(?<!\w): uma "lookbehind" negativa que verifica se há um caractere de palavra antes do hífen ou se há apenas um espaço em branco, resultando em "verdadeiro" se não houver nada
#(?!\w): uma lookahead negativa que verifica se há um caractere de palavra depois do hífen.
```

---
Será feita a tokenização dos dados, transformando as palavras em itens de lista. Para isso será utilziado o método *split()*, que colocará a vírgula como separador.

Sereá feita a tokenização de 2 colunas: descricao_local e comodidades

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
```

```python
dados[['descricao_local']]
```

Antes de fazer o mesmo para a coluna "comodidades", faremomos um tratamento dos dados pois a mesma paresenta caracteres especiais como chaves e aspas. Usaremos o *replace()*.

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
dados['comodidades']
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados['comodidades']
```
